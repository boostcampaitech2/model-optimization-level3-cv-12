```python
import optuna
import plotly
import matplotlib.pyplot as plt
```

```python
loaded_study = optuna.load_study(study_name="automl101", storage="sqlite:///search.db")
```

```python
trials= loaded_study.get_trials()
```

```python
time = []
f1=[]
for i in trials:
    if i.values != None:
        time.append(i.values[2])
        f1.append(i.values[0])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(f1, time)
plt.xlabel('F1 score')
plt.ylabel('Mean Time')
for i in range(len(time)):
    plt.annotate(str(i), (f1[i], time[i]))
```

```python
plt.savefig('score.png', dpi=300)
```
